```python
Q2. What happens when an exception is not handled? Explain with an example.

Answer-- If an exception is not handled, the program:
            . Terminates immediately
            . Display a trackback error message

        Example:
            x = 10/0
            print("hello")

        Output:  ZeroDivisionError: division by zero

     Here hello will not print because the program stops.
```

```python
Q3. Which Python statements are used to catch and handle exceptions? Explain with an example.

Answer-- Python uses:
             try -- Contains the code that may cause an error. Python first execute this block.
             except--Executes only if an eception occurs. Prevents the program from crashing.

        Example: try:
                    x = 10 / 0
                except ZeroDivisionError:
                    print("Cannot divide by zero")
```
